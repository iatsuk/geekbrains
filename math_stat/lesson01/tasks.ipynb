### Задача 1
Из колоды в 52 карты вынимают случайным образом 4 карты. Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.

Ответ. Так как туза всего 4, вероятность того, что будет вытянут один из тузов: 
$$P = \frac{4}{52} = \frac{1}{13}$$

### Задача 2
Семь человек рассаживаются наудачу на скамейке. Какова вероятность того, что рядом будут сидеть: а) два определённых человека? б) три определённых человека?

Ответ. Задача аналогична задачам по бросанию кубика.

а) Нужно, чтобы совместно произошло 2 элементарных события: сначала сел первый определённый человек из 7, а потом второй определённый человек из оставшихся 6.
$$P = \frac{1}{7} \cdot \frac{1}{6} = \frac{1}{42}$$
б) Второй случай аналогичен первому: сначала должен сесть первый определённый человек из 7, потом второй определённый человек из оставшихся 6, потом третий из оставшихся 5.
$$P = \frac{1}{7} \cdot \frac{1}{6} \cdot \frac{1}{5} = \frac{1}{210}$$

### Задача 3
Из 60 вопросов, входящих в экзаменационные билеты, студент знает 50. Какова вероятность того, что среди трёх наугад выбранных вопросов студент знает: а) все? б) два?

а) Вероятность того, что среди трёх наугад выбранных вопрос студент знает ответ на все три, равна произведению трёх событий:
* Студен достал один из 50 вопросов на которых он знал ответ из 60 вопросов в целом
* Студен достал один из 49 оставшихся вопросов на которых он знал ответ из 59 оставшихся вопросов в целом
* Студен достал один из 48 оставшихся вопросов на которых он знал ответ из 58 оставшихся вопросов в целом
$$P = \frac{50}{60} \cdot \frac{49}{59} \cdot \frac{48}{58}$$

```python
print( 'P = ', ((50 * 49 * 48) / (60 * 59 * 58)) )
```

б) Вероятность того, что среди трёх наугад выбранных вопрос студент знает ответ на два, равна произведению трёх событий:
* Студен достал один из 50 вопросов на которых он знал ответ из 60 вопросов в целом
* Студен достал один из 49 оставшихся вопросов на которых он знал ответ из 59 оставшихся вопросов в целом
* Студен достал один из 10 оставшихся вопросов на которых он знал ответ из 58 оставшихся вопросов в целом
$$P = \frac{50}{60} \cdot \frac{49}{59} \cdot \frac{10}{58}$$

```python
print( 'P = ', ((50 * 49 * 10) / (60 * 59 * 58)) )
```

### Задача 4
Бросается игральная кость. Пусть событие A - появление чётного числа, событие B - появление числа больше трёх. Являются ли эти события независимыми?

Ответ. Считаем, что $P(A) = P(B) = \dfrac{3}{6} = \dfrac{1}{2}$. Посчитаем совместную вероятность $P(AB)$. Событие $AB$ соответствует появлению чётного числа больше трёх, т.е. $P(AB) = P(A) \cdot P(B) = \dfrac{1}{4}$. Следовательно, 
$$P(A|B) = \dfrac{P(AB)}{P(B)} = \dfrac{\frac{1}{4}}{\frac{1}{2}} = \dfrac{1}{2}.$$

Поскольку $P(A|B) = P(A)$, заключаем, что события $A$ и $B$ независимы.
