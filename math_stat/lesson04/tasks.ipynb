#### Задача 1
О непрерывной равномерно распределённой случайной величине $B$ известно, что её дисперсия равна $0.2$. Можно ли найти правую границу величины $B$ и её среднее значение зная, что левая граница равна $0.5$? Если да, найдите их.

***Решение***

Дисперсия равномерного распределения равна:
$$D(B) = \frac{(b-a)^2}{12}.$$

Так как мы знаем левую границу $a = 0.5$ и значение дисперсии $D(B) = 0.2$, то мы можем подставить эти значения в формулу и вывести значение правой границы $b$:
$$0.2 = \frac{(b-0.5)^2}{12}$$
$$2.4 = (b-0.5)^2$$
$$\sqrt{2.4} = b-0.5$$
$$b = 0.5 + \sqrt{2.4}$$

Зная значения левой и правой границы равномерного распределения найдём среднее значение:
$$M(B) = \frac{a + b}{2} = \frac{0.5 + 0.5 + \sqrt{2.4}}{2} = 0.5 + \sqrt{0.6} = \frac{1}{2} + \sqrt{\frac{3}{5}}$$

С помощью правил стандартного отклонения, двух и трёх сигм посчитайте, какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:

##### • больше $182$ см?

___Решение.___ Найдём на сколько средних квадратических отклонений отклоняется рост человека, равный $182$ см для города $X$:
$$\sigma = \frac{182-174}{8} = 1$$

Так как интервал от $a - \sigma$ до $a + \sigma$ содержит около $68\%$ вероятностной массы, то оставшаяся вероятностная масса равна $100\% - 68\% = 32\%$. Исходя из того, что функция нормального распределения зеркальная, а нам нужна только правая часть вероятностной массы, то получим, что вероятность встретить взрослого человека ростом больше $182$ см равна $\frac{32}{2}\%=16\%$.
