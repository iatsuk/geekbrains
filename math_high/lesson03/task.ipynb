## Задание 1
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами.

```python
def vlen(*coords):
    s = sum([c*c for c in coords])
    return np.sqrt(s)


print('[10]:', vlen(10))
print('[10, 10]:', vlen(10, 10))
print('[10, 10, 10]:', vlen(10, 10, 10))
```

## Задание 2
Напишите код на Python, реализующий построение графика окружности.

```python
def draw_circle(x0, y0, r):
    x = np.linspace(-r, r, 101)
    y = np.sqrt(r*r - x*x)
    plt.plot(x + x0, y + y0)
    plt.plot(x + x0, -y + y0)
    plt.axis('equal')
    plt.show()


plt.figure(figsize=(5,5))
draw_circle(3, -4, 5)
```

Напишите код на Python, реализующий построение графика эллипса.

```python
def draw_ellipse(x0, y0, xr, yr):
    x = np.linspace(-xr, xr, 101)
    el_x = x*x/(xr*xr)
    y = np.sqrt((1 - el_x)*(yr*yr))
    plt.plot(x + x0, y + y0)
    plt.plot(x + x0, -y + y0)
    b = max(xr, yr)
    plt.axis('equal')
    plt.show()


plt.figure(figsize=(5,5))
draw_ellipse(5, -3, 5, 3)
```

Напишите код на Python, реализующий построение графика гиперболы.

```python
def draw_hyperbole(x0, y0, xr, yr):
    x = np.linspace(-10, 10, 101)
    el_x = x*x/(xr*xr)
    y = np.sqrt((1 + el_x)*(yr*yr))
    plt.plot(x + x0, y + y0)
    plt.plot(x + x0, -y + y0)
    b = max(xr, yr)
    plt.axis('equal')
    plt.show()


plt.figure(figsize=(5,5))
draw_hyperbole(3, 4, 2, 3)
```

## Задание 1
Нарисуйте график функции: `y(x) = k∙cos(x – a) + b` для некоторых (2-3 различных) значений параметров k, a, b.

```python
def func(x, k, a, b):
    return k*np.cos(x - a) + b


x = np.linspace(-np.pi*2, np.pi*2, 101)
plt.plot(x, func(x, 1, 1, 1), label='k = a = b = 1')
plt.plot(x, func(x, 2, 1, 1), label='k = 2')
plt.plot(x, func(x, 1, 2, 1), label='a = 2')
plt.plot(x, func(x, 1, 1, 2), label='b = 2')
plt.legend()
plt.axis('equal')
plt.show()
```

## Задание 3
Напишите код, который будет переводить полярные координаты в декартовы.

```python
def to_polar(x, y):
    p = np.sqrt(x*x + y*y)
    phi = np.arctan2(y, x)
    return p, phi


def to_dec(p, phi):
    x = p*np.cos(phi)
    y = p*np.sin(phi)
    return x, y


print(f'Полярные координаты ${to_polar(0, 2)} равны декартовым ${to_dec(2.0, np.pi/2)}')
```

Напишите код, который будет рисовать график окружности в полярных координатах.

```python
def draw_polar_circle(r):
    phi = np.linspace(0, np.pi*2, 101)
    p = [r for _ in range(len(phi))]
    plt.polar(phi, p, 'r-')
    plt.show()


plt.figure(figsize=(5,5))
draw_polar_circle(5)
```

Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.


```python
def draw_polar_line(angle, r):
    phi = [angle for _ in range(101)]
    p = np.linspace(0, r, 101)
    plt.polar(phi, p, 'r-')
    plt.show()


plt.figure(figsize=(5,5))
draw_polar_line(2, 11)
```

```python
x = np.linspace(-5, 5, 101)
plt.plot(x, x*x -1)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.show()
```

```python
from scipy.optimize import fsolve


def equations(p):
    x, y = p
    return x*x - 1 - y, np.exp(x) + x*(1 - y) - 1


print(fsolve(equations, (-2, 2)))
print(fsolve(equations, (3, 5)))
print(fsolve(equations, (5, 15)))
```
